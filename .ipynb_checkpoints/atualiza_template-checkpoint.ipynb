```python
import pandas as pd
from PIL import Image, ImageDraw, ImageFont


pd.set_option('max_colwidth', None) # Opção para exibir a largura total das colunas
```

```python
df = pd.read_csv('dataset coronavirus uberlandia.csv', index_col='Data')

df.head(10)
```

# Automatizador de templates coronavirus

```python
ultima = -3
```

```python
def AvaliaVariacao(x):
    if x[ultima]>x[ultima - 1]:
        return f"(+{(x[ultima] - x[ultima-1]) / x[ultima-1]*100:.0f}%)" 
    elif x[ultima]<x[ultima-1]:
        return f"({(x[ultima] - x[ultima-1]) / x[ultima-1]*100:.0f}%)"
    else:
        return
```

```python
img = Image.open("Template.jpeg")

d = ImageDraw.Draw(img)

fontPath = "Roboto-Black.ttf"
roboto  =  ImageFont.truetype (fontPath,45)
roboto_little  =  ImageFont.truetype (fontPath,16)

d.text((156,216), f"{df.index[ultima]}", font=roboto, fill=(254, 253, 255))

d.text((100,423), f"{df['Casos suspeitos'][ultima]:.0f}", font=roboto, fill=(0, 0, 0))
d.text((120,483), AvaliaVariacao(df['Casos suspeitos']), font=roboto_little, fill=(0, 0, 0))

d.text((350,423), f"{df['Casos confirmados'][ultima]:.0f}", font=roboto, fill=(0, 0, 0))
d.text((380,483), AvaliaVariacao(df['Casos confirmados']), font=roboto_little, fill=(0, 0, 0))

d.text((600,423), f"{df['Resultados negativos'][ultima]:.0f}", font=roboto, fill=(0, 0, 0))
d.text((640,483), AvaliaVariacao(df['Resultados negativos']), font=roboto_little, fill=(0, 0, 0))

d.text((835,413), f"{df['Internações totais'][ultima]:.0f}", font=roboto, fill=(0, 0, 0))
d.text((855,473), AvaliaVariacao(df['Internações totais']), font=roboto_little, fill=(0, 0, 0))

d.text((175,643), f"{df['Internações UTI'][ultima]:.0f}", font=roboto, fill=(0, 0, 0))
d.text((175,693), AvaliaVariacao(df['Internações UTI']), font=roboto_little, fill=(0, 0, 0))

d.text((175,719), f"{df['Internações enfermaria'][ultima]:.0f}", font=roboto, fill=(0, 0, 0))
d.text((175,763), AvaliaVariacao(df['Internações enfermaria']), font=roboto_little, fill=(0, 0, 0))

d.text((361,668), f"{df['Pacientes recuperados'][ultima]:.0f}", font=roboto, fill=(0, 0, 0))
d.text((381,728), AvaliaVariacao(df['Pacientes recuperados']), font=roboto_little, fill=(0, 0, 0))

d.text((643,668), f"{df['Óbitos Suspeitos'][ultima]:.0f}", font=roboto, fill=(0, 0, 0))
d.text((623,728), AvaliaVariacao(df['Óbitos Suspeitos']), font=roboto_little, fill=(0, 0, 0))

d.text((846,668), f"{df['Óbitos confirmados'][ultima]:.0f}", font=roboto, fill=(0, 0, 0))
d.text((846,728), AvaliaVariacao(df['Óbitos confirmados']), font=roboto_little, fill=(0, 0, 0))

d.text((105,835), f"{df['Índice de ocupação UTI'][ultima]*100:.0f}%", font=roboto, fill=(249, 241, 20))

d.text((441,914), f"{df['Leitos disponíveis'][ultima]:.0f}", font=roboto_little, fill=(0, 64, 112))

img.save(f"Boletins_Atualizados/{df.index[ultima].replace('/', '-')}.png")
```
